```python
import os

root_directory = os.getcwd() + "\\Naver\\Political"
os.chdir(root_directory)
file_list = os.listdir()
```

```python
import time
import requests
import re
import pandas as pd
from bs4 import BeautifulSoup

# 네이버 뉴스를 가져오기 위한 기본 header 저장
headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/87.0.4280.88 Safari/537.36"}

for csv_file in file_list:
    os.chdir(root_directory)
    
    if ".csv" in csv_file:
        print("Start Crawling :", csv_file)
        
        # 본문을 가져올 url 데이터 가져오기
        news_data = pd.read_csv(csv_file)
        news_urls = news_data["Url"]
        
        # 본문 데이터 저장
        start = time.time()
        d = {}
        for i, url in enumerate(news_urls):
            resp = requests.get(url, headers=headers)
            soup = BeautifulSoup(resp.text, "html.parser")
            
            # 스포츠 혹은 연예기사가 오분류 되는 경우가 있음 -> 수정될 경우 본문을 크롤링 해올 수 없으므로 예외처리
            try:
                contents = soup.find(id="articleBodyContents").get_text()
                title = url[url.find("aid=") + 4:] + ".txt"
                d[title] = contents
                if (i+1) % 100 == 0 or (i+1) == len(news_urls):
                    print("Crawling ", i+1, "/", len(news_urls))
            except:
                print("Error", i+1, "/", len(news_urls))
                
        end = time.time()
        print("Crawling Success ", round(end - start, 4))
        
        
        # 본문 문서를 저장할 폴더 지정
        directory = "C:\\Users\\cjy89\\Python_Study\\NLP\\Project_news_crawling\\Naver\\Political\\" + csv_file[:-4]
        if not os.path.isdir(directory):
            os.makedirs(directory)
        os.chdir(directory)
            
        # 파일 저장
        for title in d:
            f = open(title, "wt", encoding="utf-8-sig")
            f.write(d[title].strip())
            f.close()
        print("Saving File Success :", csv_file)
        
    print("──────────────────────")
```
